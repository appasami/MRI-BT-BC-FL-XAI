```python
import os
import pandas as pd

# Define dataset paths with names
dataset_paths = {
    "D1": "/kaggle/input/d/jjprotube/brain-mri-images-for-brain-tumor-detection/brain_tumor_dataset",
    "D2": "/kaggle/input/brain-mri-scan-dataset/dataset/dataset/train",
    "D3": "/kaggle/input/brain-mri-scan-dataset/dataset/dataset/valid",
    "D4": "/kaggle/input/brain-tumor/2 classes",
    "D5": "/kaggle/input/brain-tumor-detection/dataset",
    "D6": "/kaggle/input/brain-tumor-mri-yes-or-no/Brain (y-n)/Training",
    "D7": "/kaggle/input/brain-tumor-mri-dataset/Brain_Tumor_Detection",
    "D8": "/kaggle/input/brain-tumor-mri-datasets/Brain_Tumor_Datasets/train",
    "D9": "/kaggle/input/brain-tumor-mri-datasets/Brain_Tumor_Datasets/test",
    "D10": "/kaggle/input/mri-brain-tumor-yes-no",
    "D11": "/kaggle/input/mri-mini-brain-tumor-binary-class-yes-no",
    "D12": "/kaggle/input/test-1/brain-mri-images-for-brain-tumor-detection/New Folder/brain_tumor_dataset",
}

# Count images in each dataset
total_counts = {}
for name, path in dataset_paths.items():
    count_yes, count_no = 0, 0
    for label in ['yes', 'no']:
        subdir = os.path.join(path, label)
        if os.path.exists(subdir):
            count = len([
                file for file in os.listdir(subdir)
                if os.path.isfile(os.path.join(subdir, file))
            ])
            if label == 'yes':
                count_yes = count
            else:
                count_no = count
    total_counts[name] = {'path': path, 'yes': count_yes, 'no': count_no}

# Create DataFrame
rows = []
for name, info in total_counts.items():
    rows.append({
        "Name": name,
        "Dataset": info["path"],
        "YES": info["yes"],
        "NO": info["no"],
        "Total": info["yes"] + info["no"]
    })

df = pd.DataFrame(rows)
print(df.to_markdown(index=False))
```

```python
import os
import tensorflow as tf
import numpy as np
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras import layers, models, optimizers
import pandas as pd

# Create directories to save plots
os.makedirs("visualizations/training_history", exist_ok=True)
os.makedirs("visualizations/confusion_matrices", exist_ok=True)

tf.random.set_seed(42)
np.random.seed(42)

epochs = 32  
batch_size =32
lr = 0.001
IMG_HEIGHT = 128
IMG_WIDTH = 128
CHANNELS = 3

cross_test_paths = [
    "/kaggle/input/d/jjprotube/brain-mri-images-for-brain-tumor-detection/brain_tumor_dataset",
    "/kaggle/input/brain-mri-scan-dataset/dataset/dataset/train",
    "/kaggle/input/brain-mri-scan-dataset/dataset/dataset/valid",
    "/kaggle/input/brain-tumor/2 classes",
    "/kaggle/input/brain-tumor-detection/dataset",
    "/kaggle/input/brain-tumor-mri-yes-or-no/Brain (y-n)/Training",
    "/kaggle/input/brain-tumor-mri-dataset/Brain_Tumor_Detection",
    "/kaggle/input/brain-tumor-mri-datasets/Brain_Tumor_Datasets/train",
    "/kaggle/input/brain-tumor-mri-datasets/Brain_Tumor_Datasets/test",
    "/kaggle/input/mri-brain-tumor-yes-no",
    "/kaggle/input/mri-mini-brain-tumor-binary-class-yes-no",
    "/kaggle/input/test-1/brain-mri-images-for-brain-tumor-detection/New Folder/brain_tumor_dataset",
]

dataset_names = [f"d{i+1}" for i in range(len(cross_test_paths))]

def load_and_preprocess_dataset(data_path, shuffle=True):
    if not os.path.exists(data_path):
        print(f"Warning: Dataset path '{data_path}' does not exist. Skipping.")
        return None, 0
    
    print(f"Attempting to load data from: {data_path}")
    try:
        dataset = tf.keras.utils.image_dataset_from_directory(
            data_path,
            labels='inferred',
            label_mode='binary',
            image_size=(IMG_HEIGHT, IMG_WIDTH),
            interpolation='nearest',
            batch_size=batch_size,
            shuffle=shuffle,
        )
        num_samples = len(dataset.file_paths)
        dataset = dataset.map(lambda x, y: (tf.cast(x, tf.float32) / 255.0, y))
        print(f"Successfully loaded data from: {data_path} ({num_samples} samples)")
        return dataset, num_samples
    except Exception as e:
        print(f"Error loading dataset from {data_path}: {e}")
        return None, 0

def build_cnn_model():
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, CHANNELS),
                      padding='same', name='conv2d_1'),
        layers.MaxPooling2D((2, 2), name='max_pooling2d_1'),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same', name='conv2d_2'),
        layers.MaxPooling2D((2, 2), name='max_pooling2d_2'),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same', name='conv2d_3'),
        layers.MaxPooling2D((2, 2), name='max_pooling2d_3'),
        layers.Flatten(name='flatten'),
        layers.Dense(128, activation='relu', name='dense_1'),
        layers.Dropout(0.5, name='dropout_1'),
        layers.Dense(1, activation='sigmoid', name='output_layer')
    ])
    optimizer = optimizers.Adam(learning_rate=lr)
    model.compile(optimizer=optimizer,
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

def plot_training_history(history, train_name):
    """Plots and saves the training accuracy and loss curves."""
    plt.figure(figsize=(12, 5))
    
    # Plot training accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.title(f'Training Accuracy (Trained on {train_name})')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # Plot training loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.title(f'Training Loss (Trained on {train_name})')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.savefig(f"visualizations/training_history/history_{train_name}.png")
    plt.close()

def plot_confusion_matrix(true_labels, predicted_classes, train_name, test_name):
    """Plots and saves the confusion matrix."""
    cm = confusion_matrix(true_labels, predicted_classes)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['no', 'yes'], yticklabels=['no', 'yes'])
    plt.title(f'Confusion Matrix\nTrained on {train_name}, Tested on {test_name}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.savefig(f"visualizations/confusion_matrices/cm_trained_{train_name}_tested_{test_name}.png")
    plt.close()

all_results = {}
accuracy_matrix = np.zeros((len(dataset_names), len(dataset_names)))
dataset_sizes = {}

# First, get all dataset sizes
for i, path in enumerate(cross_test_paths):
    name = dataset_names[i]
    _, num_samples = load_and_preprocess_dataset(path, shuffle=False)
    dataset_sizes[name] = num_samples

for i, train_path in enumerate(cross_test_paths):
    train_name = dataset_names[i]
    all_results[train_name] = {}
    print(f"\n============================================================")
    print(f"--- Training Model with Dataset: {train_name} ({train_path}) ---")
    print(f"============================================================")

    train_ds, _ = load_and_preprocess_dataset(train_path, shuffle=True)
    if train_ds is None:
        print(f"Skipping training with {train_name} due to data loading error or non-existent path.")
        accuracy_matrix[i, :] = np.nan
        continue

    model = build_cnn_model()
    
    print(f"Starting training for model on {train_name} for {epochs} epochs...")
    try:
        history = model.fit(
            train_ds,
            epochs=epochs,
            verbose=1
        )
        print(f"Training with {train_name} completed successfully.")
        plot_training_history(history, train_name)
    except Exception as e:
        print(f"Error during training with {train_name}: {e}")
        accuracy_matrix[i, :] = np.nan
        continue

    for j, test_path in enumerate(cross_test_paths):
        test_name = dataset_names[j]
        print(f"\n--- Testing Model (trained on {train_name}) with Dataset: {test_name} ({test_path}) ---")

        test_ds, _ = load_and_preprocess_dataset(test_path, shuffle=False)
        if test_ds is None:
            all_results[train_name][test_name] = {'accuracy': np.nan, 'report': "Test dataset not loaded."}
            accuracy_matrix[i, j] = np.nan
            print(f"Skipping testing with {test_name} due to data loading error or non-existent path.")
            continue

        all_true_labels = []
        all_predicted_probs = []

        try:
            print(f"Collecting predictions for {test_name}...")
            for images, labels in test_ds:
                all_true_labels.extend(labels.numpy())
                predictions = model.predict(images, verbose=0)
                all_predicted_probs.extend(predictions.flatten())

            true_labels = np.array(all_true_labels)
            predicted_probs = np.array(all_predicted_probs)
            predicted_classes = (predicted_probs > 0.5).astype(int)

            if true_labels.size == 0:
                accuracy = 0.0
                report = "No samples in test dataset to evaluate."
                print(f"Warning: No samples found in {test_name}. Accuracy set to 0.0.")
            else:
                accuracy = accuracy_score(true_labels, predicted_classes)
                report = classification_report(true_labels, predicted_classes, digits=4, target_names=['no', 'yes'], zero_division=0, output_dict=True)
                plot_confusion_matrix(true_labels, predicted_classes, train_name, test_name)

            print(f"Accuracy on {test_name}: {accuracy:.4f}")
            print("Classification Report:\n", classification_report(true_labels, predicted_classes,  digits=4, target_names=['no', 'yes'], zero_division=0))

            all_results[train_name][test_name] = {'accuracy': accuracy, 'report': report}
            accuracy_matrix[i, j] = accuracy

        except Exception as e:
            print(f"Error during testing model (trained on {train_name}) with {test_name}: {e}")
            all_results[train_name][test_name] = {'accuracy': np.nan, 'report': f"Error during evaluation: {e}"}
            accuracy_matrix[i, j] = np.nan

# Convert accuracy matrix to a DataFrame for easier analysis
accuracy_df = pd.DataFrame(accuracy_matrix, index=dataset_names, columns=dataset_names)

print("\n\n====================================================")
print("---           ADDITIONAL VISUALIZATIONS          ---")
print("====================================================\n")

# --- 1. Main Accuracy Heatmap ---
print("--- Generating Cross-Dataset Accuracy Heatmap ---")
plt.figure(figsize=(14, 12))
sns.heatmap(accuracy_df, annot=True, fmt=".4f", cmap="viridis", linewidths=.5, linecolor='black')
plt.title('Cross-Dataset Classification Accuracy Heatmap\n(Rows: Trained On, Columns: Tested On)', fontsize=16)
plt.xlabel('Tested On Dataset', fontsize=14)
plt.ylabel('Trained On Dataset', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig("accuracy_heatmap.png")
plt.show()


# --- 2. Bar Chart of Average "Trainer" Accuracy ---
print("\n--- Generating Average 'Trainer' Accuracy Bar Chart ---")
avg_trainer_accuracy = accuracy_df.mean(axis=1).sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x=avg_trainer_accuracy.values, y=avg_trainer_accuracy.index, palette="plasma", orient='h')
plt.xlabel('Average Test Accuracy', fontsize=12)
plt.ylabel('Training Dataset', fontsize=12)
plt.title('Average Model Performance by Training Dataset (Robustness)', fontsize=16)
plt.xticks(rotation=45)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("average_trainer_accuracy.png")
plt.show()

# --- 3. Bar Chart of Average "Tester" Accuracy ---
print("\n--- Generating Average 'Tester' Accuracy Bar Chart ---")
avg_tester_accuracy = accuracy_df.mean(axis=0).sort_values(ascending=True)
plt.figure(figsize=(12, 8))
sns.barplot(x=avg_tester_accuracy.values, y=avg_tester_accuracy.index, palette="cividis", orient='h')
plt.xlabel('Average Accuracy Achieved by Models', fontsize=12)
plt.ylabel('Testing Dataset', fontsize=12)
plt.title('Average Dataset Score When Used for Testing (Difficulty)', fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("average_tester_accuracy.png")
plt.show()


# --- 4. Box Plot of Accuracy Distributions ---
print("\n--- Generating Accuracy Distribution Box Plot ---")
plt.figure(figsize=(15, 8))
sns.boxplot(data=accuracy_df.T, palette="Set3", orient='v')
plt.ylabel('Accuracy Distribution', fontsize=12)
plt.xlabel('Training Dataset', fontsize=12)
plt.title('Distribution of Test Accuracies for Each Training Dataset', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("accuracy_distribution_boxplot.png")
plt.show()


# --- 5. Bar Chart of Dataset Sizes ---
print("\n--- Generating Dataset Size Bar Chart ---")
dataset_sizes_df = pd.Series(dataset_sizes).sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x=dataset_sizes_df.values, y=dataset_sizes_df.index, palette="rocket", orient='h')
plt.xlabel('Number of Samples', fontsize=12)
plt.ylabel('Dataset', fontsize=12)
plt.title('Number of Samples per Dataset', fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("dataset_sizes.png")
plt.show()

print("\n============================================")
print("--- Detailed Results Summary ---")
print("============================================")
for train_name, test_results in all_results.items():
    print(f"\nModel trained on {train_name}:")
    for test_name, metrics in test_results.items():
        if isinstance(metrics['accuracy'], float) and not np.isnan(metrics['accuracy']):
            print(f"  Tested on {test_name}: Accuracy = {metrics['accuracy']:.4f}")
        else:
            print(f"  Tested on {test_name}: {metrics['report']}")
```
